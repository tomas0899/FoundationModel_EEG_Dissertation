## print all T0 and check time ranges

```python
mat_files = sorted([
    f for f in os.listdir(folder_path)
    if f.lower().endswith(".mat") and not f.startswith(".")
])

print("Number of .mat files:", len(mat_files))
```

```python
t0_results = []
errors = []

for i, filename in enumerate(mat_files):
    file_path = folder_path / filename

    try:
        data_mat = loadmat(file_path)
        hdr = data_mat["hdr"]

        # Extract T0 from original header
        t0_raw = hdr["orig"][0, 0]["T0"][0, 0][0]

        t0_dt = datetime(
            int(t0_raw[0]), int(t0_raw[1]), int(t0_raw[2]),
            int(t0_raw[3]), int(t0_raw[4]), int(t0_raw[5])
        )

        t0_results.append({
            "list_idx": i,
            "file": filename,
            "T0": t0_dt
        })

    except Exception as e:
        errors.append({
            "list_idx": i,
            "file": filename,
            "error": str(e)
        })

df_T0 = pd.DataFrame(t0_results)
df_T0 = df_T0.sort_values("T0").reset_index(drop=True)

print("Successfully processed:", len(df_T0))
print("Errors:", len(errors))

display(df_T0)
```

```python
t0_raw_results = []
errors = []

for i, filename in enumerate(mat_files):
    file_path = folder_path / filename

    try:
        data_mat = loadmat(file_path)
        hdr = data_mat["hdr"]

        # Raw T0 exactly as stored in the .mat structure
        t0_raw = hdr["orig"][0, 0]["T0"][0, 0][0]

        t0_raw_results.append({
            "list_idx": i,
            "file": filename,
            "T0_raw": t0_raw
        })

    except Exception as e:
        errors.append({
            "list_idx": i,
            "file": filename,
            "error": str(e)
        })

df_T0_raw = pd.DataFrame(t0_raw_results)

print("Successfully processed:", len(df_T0_raw))
print("Errors:", len(errors))

display(df_T0_raw)
```
