## Function guide: `process_eeg_mat_files_1_1`

### Purpose

The function `process_eeg_mat_files_1_1()` scans a folder containing EEG `.mat` files and extracts temporal information from each recording.

Its main goal is to generate a structured overview of the recording timeline, including:

- the file name,
- the recording start time (`T0`),
- the recording end time (`TF`),
- the recording duration,
- and the time gaps between consecutive recordings.

This function is useful as an initial quality-control step.
